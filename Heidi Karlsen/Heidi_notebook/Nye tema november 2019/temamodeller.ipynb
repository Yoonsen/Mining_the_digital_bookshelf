```python
korpus = nb.book_corpus(period=(1830, 1880), limit = 10000)
korpus
```

```python
aggs = []
step = 300
urns = nb.pure_urn(korpus.urn)
for i in range(0, len(korpus), step):
    aggs.append(nb.frame(nb.get_aggregated_corpus(urns[i:i+step]), str(i)))
```

```python
aggs_korpus = pd.concat([a for a in aggs], axis=1).fillna(0)
korpus_agg = aggs_korpus.sum(axis=1)
```
