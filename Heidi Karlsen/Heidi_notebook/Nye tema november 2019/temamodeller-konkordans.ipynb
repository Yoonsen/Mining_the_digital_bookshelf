```python
import dhlab.nbtext as nb
import dhlab.module_update as mu
import pandas as pd
import json
```

Søkeordene ligger i temadefinisjonene, ikke alle fins på 1800-tallet, men noen er der, med varierende frekvens

```python
#temaene ble lagret med annen notebook - Heidis definisjoner_0_november
with open('heidis_temadefinisjoner_november.json') as f:
    first = json.load(f)
```

```python
wordbags = dict()
for f in first:
    for wb in first[f]:
        wordbags[wb] = first[f][wb]
```

### Ordsekkfrekvenser

Få en oversikt over frekvensen til ordene, merk at det kan godt hende at ordene ikke fins i selve korpuset. Referansefrekvensene er en forkortet fil med ord som har frekvens over 5 totalt, mest for å ned størrelsen på selve korpuset.

```python
# leser inn korpus_agg. Pass på å få den med om du flytter på notebøkene.

korpus_agg = pd.read_json('Referanse_bøker_1830_1880_f5.json', typ='series')
```

Benytt cellen under (kopier og bruk på nytt) for å sjekke frekvensen av ordsekkene i korpuset - bare bytt ut identifikatoren i rødt med en annen. Det gir deg et blikk på hva som fins. Laveste frekvens vil være 5. Ord med lavere frekvens er ikke med.

```python
korpus_agg.loc[[x for x in wordbags['Kvindeforening_'] if x in korpus_agg.index]]
```

### Korpusdefinisjon

Korpuset er bøkene mellom 1830 og 1880

```python
korpus = nb.book_corpus(period=(1830, 1880), limit = 10000)
korpus
```

### Konkordanser 

Benytt koden under for å finne konkordanser med lenke til Bokhylla. 

```python
konkword = lambda x: nb.urn_concordance(word = x, urns = nb.refine_book_urn(urns = nb.pure_urn(korpus.urn), words = [x]), before=12, after=12, size=20)
konklist = lambda x: nb.urn_concordance(word = x, urns = nb.refine_book_urn(urns = nb.pure_urn(korpus.urn), words = x), before=12, after=12, size=20)
```

Kommandoen konkword tar ett eneste ord, mens konklist tar en liste, for eksempel en ordsekk. Se de to eksemplene under.

```python
konkword('Kvindeforening')
```

```python
konklist(wordbags['Kvindeforening_'])
```
