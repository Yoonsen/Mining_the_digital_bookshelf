```python
def get_paras(filename):
    """Get a list of paragraphs"""
    with open(filename, encoding='utf-8') as f:
        content = f.read()
    soup = BeautifulSoup(content, 'html.parser')
    text = []
    for para in soup.find_all('p'):
        text.append(tokenize(para.text))
    return text
```
