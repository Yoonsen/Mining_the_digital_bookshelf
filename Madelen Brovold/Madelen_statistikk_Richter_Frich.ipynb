# Bygg korpus og analysér

Hent URN-er fra tekst eller metadata. URNene representer korpuset, eller en definisjon av korpuset. Korpuset bygges med ordene som metadata lagt inn i variabelen `words`. Korpusets URNer ligger i variabelen `korpusår`.

```python
words = """ jøde, jøden, jøder, jødene, jøderne, jødisk, jødiske""".split(', ')
words = words + [x.capitalize() for x in words]
```

```python
Richterkorpus_def = nb.get_urn({'author':'%frich%richter%', 'limit':200})
len(Richterkorpus_def)
```

```python
Richterkorpus_def
```

```python
aggs = nb.get_aggregated_corpus(Richterkorpus_def)
```

```python
aggs
```

## Kollokasjoner 

Her kjøres kollokasjoner for ord med forskjellige krav til frekvens. 

### Kollokasjoner for n=5

For å være med i kollokasjonene kreves at ordet forekommer minst 6 ganger i hele korpuset

```python
colls = nb.urn_coll_words(words, urns=Richterkorpus_def)
```

```python
colls10 = nb.urn_coll_words(words, urns=Richterkorpus_def, before=10, after=10)
```

```python
nb.normalize_corpus_dataframe(colls10)
```

```python
nb.normalize_corpus_dataframe(colls)
nb.normalize_corpus_dataframe(aggs)
```
